```julia
# Special Functions
# Delta Distribution
delta(x) = 1/((1e-3)*√(2π))*exp(-1/2*(x/(1e-3))^2)
# Heaviside Function
H(x) = 0.5 * (sign(x)+1)
# Triangle Function
Λ(x) = H(x+1) * H(-(x-1)) * (1-abs(x))
# Sign Function
sgn(x) = sign(x)
# Triangle Function
rect(x) = H(x+0.5) * H(-(x-0.5));
```

## Gauß-Funktion

Wir haben folgendes Ausdruck
$$f(x,y)=e^{-\pi\left(a^2x^2+b^2y^2\right)}$$
Und versuchen eine Fourier-Transformation zu finden
$$\mathcal{F}\{f(x,y)\}=g(u,v)=\frac{1}{2\pi}\int_{-\infty}^{\infty}\int_{-\infty}^{\infty}f(x,y)e^{-i(ux+vy)}dxdy$$
$$\Rightarrow g(u,v)=\frac{1}{2\pi}\int_{-\infty}^{\infty}\int_{-\infty}^{\infty}e^{-\pi\left(a^2x^2+b^2y^2\right)}e^{-i(ux+vy)}dxdy$$
$$\Rightarrow g(u,v)=\frac{1}{2\pi}\int_{-\infty}^{\infty}e^{-\pi b^2y^2 ivy}\int_{-\infty}^{\infty}e^{-\pi a^2x^2-iux}dxdy$$
Wir wissen schon, dass
$$\int_{-\infty}^{\infty}e^{-ax^2+bx+c}dx=e^{\frac{b^2}{4a}+c}\sqrt{\frac{\pi}{a}}$$
Dann haben wir
$$\Rightarrow g(u,v)=\frac{1}{2\pi}\int_{-\infty}^{\infty}e^{\frac{(-iu)^2}{4\pi a^2}}\sqrt{\frac{\pi}{\pi a^2}}e^{-\pi b^2y^2 ivy}dy$$
$$\Rightarrow g(u,v)=\frac{1}{2\pi}e^{\frac{-u^2}{4\pi a^2}}\frac{1}{|a|}e^{\frac{(-iv)^2}{4\pi b^2}}\sqrt{\frac{\pi}{\pi b^2}}$$
$$\Rightarrow g(u,v)=\frac{1}{2\pi}e^{\frac{-u^2}{4\pi a^2}}\frac{1}{|a|}e^{\frac{-v^2}{4\pi b^2}}\frac{1}{|b|}$$
$$\Rightarrow g(u,v)=\frac{1}{2\pi|ab|}e^{-\frac{u^2}{4\pi a^2}-\frac{v^2}{4\pi b^2}}$$

```julia
# Gaussian
gr()
x = -3:0.01:3
y = -3:0.01:3
a = 1
b = 1
fun(x,y) = exp(-π*(a^2*x^2+b^2*y^2))
Fun(x,y) = 1/(2π*abs(a*b))*exp(-(1/4π)*(a^(-2)*x^2+b^(-2)*y^2))
p1 = surface(x,y,fun,title = "f(x)")
p2 = surface(x,y,Fun,title = "F(x)")
plot(p1,p2)
```

## Dreiecksfunktion

```julia
# Triangle Function
gr()
x = -2:0.01:2
y = -2:0.01:2
a = 1/2
b = 1/2
fun(x,y) = Λ(a*x)*Λ(b*y)
Fun(x,y) = 1/(abs(a*b))*(sinc(x/a))^2*(sinc(y/b))^2
p1 = surface(x,y,fun,title = "f(x)")
p2 = surface(x,y,Fun,title = "F(x)")
plot(p1,p2)
```

## Rechtecksfunktion

```julia
# Rectangle Function
gr()
x = -4:0.01:4
y = -4:0.01:4
a = 1/2
b = 1/2
fun(x,y) = rect(a*x)*rect(b*y)
Fun(x,y) = 1/(abs(a*b))*(sinc(x/a))*(sinc(y/b))
p1 = surface(x,y,fun,title = "f(x)")
p2 = surface(x,y,Fun,title = "F(x)")
plot(p1,p2)
```

## Kreisfunktion

```julia
# Circle Function
using SpecialFunctions
circ(x,y) = H(-(√(x^2+y^2)-1))

gr()
x = -2:0.01:2
y = -2:0.01:2
a = 1/2
b = 1/2
fun(x,y) = circ(x,y)
Fun(x,y) = besselj(1,2*π*√(x^2+y^2))/√(x^2+y^2)
# Bessel-Funktion erster Gattung 1er Ordnung
p1 = surface(x,y,fun,title = "f(x)")
p2 = surface(x,y,Fun,title = "F(x)")
plot(p1,p2)
```
